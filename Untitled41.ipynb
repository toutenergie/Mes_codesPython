```python
import re
import os

# Chemin du fichier sur Google Drive
chemin_fichier = '/content/data_loading_with_docling_loader (1).txt'

# Charger et afficher le contenu du fichier, assigning it to the variable 'contenu'
try:
    with open(chemin_fichier, "r", encoding="utf-8") as file:
        contenu = file.read() #This line assigns the content of the file to the variable 'contenu'
    print(f"Contenu du fichier {chemin_fichier} :\n")
    print(contenu)
except FileNotFoundError:
    print(f"Le fichier n'a pas été trouvé : {chemin_fichier}")
    contenu = "" # Assign an empty string to 'contenu' if the file is not found to avoid the NameError
except Exception as e:
    print(f"Une erreur s'est produite : {e}")
    contenu = "" # Assign an empty string to 'contenu' if any other error occurs

# Texte à découper
texte = contenu
# Fonction de découpage
def decouper_texte_en_chunks(texte):
    # Séparer par les titres principaux (titres en majuscules avec indentation)
    chunks = re.split(r"(?=\n[A-Z ]+\n)", texte)
    # Supprimer les espaces inutiles autour des chunks
    chunks = [chunk.strip() for chunk in chunks if chunk.strip()]
    return chunks

# Découper le texte
chunks = decouper_texte_en_chunks(texte)

# Afficher les chunks avec leur index
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}:\n{chunk}\n{'-'*80}")
```

```python
from transformers import AutoModel, AutoTokenizer
import torch
import numpy as np

# Initialiser le modèle et le tokenizer
model = AutoModel.from_pretrained("jinaai/jina-embeddings-v3", trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained("jinaai/jina-embeddings-v3", trust_remote_code=True)

# Exemple de chunks (reprenez les chunks découpés précédemment)
chunks = [
    "PRÉ DEMANDE ET PRISE DE RENDEZ-VOUS POUR L'ÉTABLISSEMENT DU PASSEPORT ORDINAIRE BIOMÉTRIQUE",
    "Dans la mise en œuvre de la dématérialisation des services publics...",
    "La pré demande se fera désormais en ligne 24h/24 à partir du 03 mai 2021...",
]

# Convertir les chunks en embeddings
def convertir_chunks_en_embeddings(chunks):
    embeddings = []
    for chunk in chunks:
        # Tokenisation
        tokens = tokenizer(chunk, return_tensors="pt", padding=True, truncation=True, max_length=512)
        # Obtenir les embeddings
        with torch.no_grad():
            outputs = model(**tokens)
        # Utiliser le vecteur [CLS] comme embedding pour le chunk
        # Convert the tensor to float32 before calling .numpy()
        embedding = outputs.last_hidden_state[:, 0, :].squeeze(0).type(torch.float32).numpy()
        embeddings.append(embedding)
    return embeddings

# Appeler la fonction pour convertir les chunks
embeddings = convertir_chunks_en_embeddings(chunks)

# Afficher la dimension et un exemple d'embedding
print(f"Nombre d'embeddings générés : {len(embeddings)}")
print(f"Dimension d'un embedding : {embeddings[0].shape}")
print(f"Exemple d'embedding :\n{embeddings[0]}")
```

```python
from transformers import AutoTokenizer, AutoModel
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import torch

# Charger le modèle et le tokenizer
model_name = "sentence-transformers/all-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Étape 1 : Générer les embeddings pour les chunks
def generate_embeddings(chunks):
    embeddings = []
    for chunk in chunks:
        # Tokenisation et génération d'embeddings
        tokens = tokenizer(chunk, return_tensors="pt", padding=True, truncation=True)
        with torch.no_grad():
            outputs = model(**tokens)
        embedding = outputs.last_hidden_state[:, 0, :].squeeze(0).numpy()
        embeddings.append(embedding)
    return np.array(embeddings)

# Étape 2 : Trouver les Top N chunks pertinents
def find_top_chunks(question, embeddings, chunks, top_n=3):
    # Générer l'embedding de la question
    tokens = tokenizer(question, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        question_embedding = model(**tokens).last_hidden_state[:, 0, :].squeeze(0).numpy()

    # Calculer les similarités cosinus
    similarities = cosine_similarity([question_embedding], embeddings)[0]

    # Trouver les indices des chunks les plus similaires
    top_indices = similarities.argsort()[-top_n:][::-1]

    # Retourner les chunks pertinents et leurs indices
    top_chunks = [(idx, chunks[idx], similarities[idx]) for idx in top_indices]
    return top_chunks

# Étape 3 : Identifier le chunk contenant la réponse exacte
def extract_best_chunk(question, top_chunks):
    # Parcourir les chunks pertinents pour trouver la réponse
    for idx, chunk, score in top_chunks:
        if question.lower() in chunk.lower():  # Recherche approximative (modulable si nécessaire)
            return idx, chunk, score
    # Si aucun chunk ne contient directement la réponse, retourner le chunk le plus pertinent
    return top_chunks[0]

# Pipeline Principal
def pipeline(question, chunks, top_n=3):
    """
    Pipeline pour trouver le chunk contenant la réponse à une question donnée.
    """
    # Étape 1 : Générer les embeddings des chunks
    embeddings = generate_embeddings(chunks)

    # Étape 2 : Trouver les Top N chunks
    top_chunks = find_top_chunks(question, embeddings, chunks, top_n)
    print("\nTop N Chunks Pertinents :")
    for idx, chunk, score in top_chunks:
        print(f"Index : {idx}, Score : {score:.2f}, Chunk : {chunk}")

    # Étape 3 : Identifier le chunk contenant la réponse exacte
    idx, best_chunk, score = extract_best_chunk(question, top_chunks)
    print("\nChunk Contenant la Réponse :")
    print(f"Index : {idx}, Score : {score:.2f}, Chunk : {best_chunk}")

    # Retourner le chunk contenant la réponse
    return idx, best_chunk
```

```python
# Exemple de chunks
chunks = [
    "Les documents nécessaires pour un passeport sont : une copie d'acte de naissance et une photo.",
    "Le délai pour la délivrance des passeports biométriques ordinaires est d’une semaine.",
    "Dans la mise en œuvre de la dématérialisation des services publics..."
]

# Exemple de question
question = "Quels sont les documents nécessaires pour obtenir un passeport ?"

# Exécuter le pipeline
index, response_chunk = pipeline(question, chunks, top_n=3)

# Résultat final
print("\nRésultat Final :")
print(f"Index du chunk contenant la réponse : {index}")
print(f"Chunk : {response_chunk}")
```

```python
# Définir le template
prompt_template = """
### Instruction
{instruction}

### Contexte
{contexte}

### Question
{question}

### Style
{style}

### Réponse
{reponse}
"""

# Fonction pour générer un prompt
def generer_prompt(question, contexte, reponse, instruction, style):
    # Remplir le template avec les entrées dynamiques
    filled_prompt = prompt_template.format(
        question=question,
        contexte=contexte,
        reponse=reponse,
        instruction=instruction,
        style=style
    )
    return filled_prompt

# Exemple de données
question = "Quelle est la durée pour obtenir un acte de naissance sécurisé ?"
contexte = """
La durée estimée pour obtenir un acte de naissance sécurisé est de 48 heures.
Le coût est de 1000 FCFA. L'Agence Nationale d’Identification des Personnes est responsable de ce service.
"""
reponse = "La durée pour obtenir un acte de naissance sécurisé est de 48 heures."
instruction = "Répondez à la question de manière concise et professionnelle en utilisant uniquement les informations du contexte."
style = "Ton formel et professionnel."

# Générer le prompt
generated_prompt = generer_prompt(question, contexte, reponse, instruction, style)

# Afficher le prompt généré
print(generated_prompt)
```
